```python
#Removing extra white spaces in the description column 
import re
df['Description']=df['Description'].astype('str').apply(lambda x: re.sub(r'\s+', ' ', x).strip())
```

```python
#InvoiceNo column contains records that are cancelled. Therefore the datatype is 'object'
df['InvoiceNo']=df['InvoiceNo'].astype('str')
df=df[~df['InvoiceNo'].str.contains('C')]
```

```python
# focusing on UK
#grouping the data by UK, alternatively for ease of analysis, one can focus on different geographies for better inference

df_grouped = (df[df['Country']=='United Kingdom']
             .groupby(['InvoiceNo','Description'])['Quantity']
             .sum().unstack().fillna(0).reset_index()
             .set_index('InvoiceNo'))
```

```python
df_grouped.head()
```

```python
#since the apriori algorithm expects the one-hot encoded format,we convert the quantities into 1 for positive values and o for values less than 0.
```

```python
def encoding(x):
    if x<=0:
        return 0
    if x >=1:
        return 1
    
df_grouped=df_grouped.applymap(encoding)
```

```python
#Dropping postage because we do not need them
df_grouped.drop('POSTAGE',inplace=True, axis=1)
```

```python
#Model Building
frq_itemsets = apriori(df_grouped, min_support=0.03, use_colnames=True)
```

```python
#generating rules
rules=association_rules(frq_itemsets, metric='confidence', min_threshold=0.1)
```

```python
rules.sort_values(by='lift',ascending=False)
```

```python
rules.to_csv('association_rules.csv', index=False)
```

```python
# Function to evaluate different min_threshold values

def evaluate_min_threshold(frq_itemsets, threshold_values):
    results = []
    for min_threshold in threshold_values:
        rules = association_rules(frq_itemsets, metric='confidence', min_threshold=min_threshold)
        results.append((min_threshold, len(rules)))
    return pd.DataFrame(results, columns=['min_threshold', 'num_rules'])


chosen_support = 0.03
frq_itemsets = apriori(df_grouped, min_support=chosen_support, use_colnames=True)

# Evaluate different min_threshold values
threshold_values = np.arange(0.1, 1.0, 0.1)
threshold_results = evaluate_min_threshold(frq_itemsets, threshold_values)

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(threshold_results['min_threshold'], threshold_results['num_rules'], marker='o')
plt.xlabel('min_threshold (Confidence)')
plt.ylabel('Number of Rules')
plt.title('Effect of min_threshold on Rules')
plt.show()
```

```python
#visualizing the rules
import networkx as nx


def plot_rules(rules):
    G = nx.DiGraph()
    for _, rule in rules.iterrows():
        for antecedent in rule['antecedents']:
            for consequent in rule['consequents']:
                G.add_edge(antecedent, consequent, weight=rule['lift'])
    pos = nx.spring_layout(G)
    plt.figure(figsize=(13, 10))
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=12, font_weight='bold')
    plt.show()

plot_rules(rules.head(20))
```
